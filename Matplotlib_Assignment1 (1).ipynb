```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

## Q1: Create two charts using list and dataframe.

```python
# Our List 
Data = [["Australia", 2500, 1980,38000],["Bangladesh", 1000, 1999,40000],["England", 2000, 2019,45000],["India", 3000, 2021,55000],["Srilanka", 1500, 2002,88000]]

# Our DataFrame 
DFrame = pd.DataFrame(Data, columns=["Team","Rank_Points", "Year", 'Audience Number'])

# plotting the dataframe
DFrame.plot(x="Team", y=["Rank_Points","Year"], kind="line", figsize=(10, 9))

```


## Create pie chart with equal explode values.


```python
# Data Visualization using Python
# Pie Chart
Teams = ['Bangladish','England','Australia','India','Srilanka']
RP= [1000,2000,2500,3000,1500]
DFrame.groupby('Team').sum().sort_values(by='Rank_Points',ascending = False)
colors = ['#ff9999','#66b3ff','#99ff99','#ffcc99']
plt.pie(RP,labels=Teams, autopct='%1.2f%%', colors =colors,  startangle=180, radius=1.5)
plt.show()



```

## Q3: Create a scatter chart with circle marker, alpha = 0.4, and size = 50.

```python
plt.scatter(DFrame['Year'],DFrame['Audience Number'], alpha = 0.4, s=50) 

plt.title('The relation between Year and Audience Number: ',fontsize = 15)
plt.xlabel('Year',fontsize = 20)
plt.ylabel('Audience Number',fontsize = 20)

plt.show()
```

## Q4: Add the following to your chart:
- title placed in the left and has a red color
- x and y label with font size = 16

```python
plt.scatter(DFrame['Year'],DFrame['Audience Number'], alpha = 0.4, s=50) 

plt.title('The relation between Year and Audience Number: ' ,color = 'red' , fontsize = 13, loc='left' )
plt.xlabel('Year',fontsize = 16)
plt.ylabel('Audience Number',fontsize = 16)

plt.show()
```

## Q6: Create subplots with three charts and figsize=20,10.

```python
plt.figure(figsize=[20,10])
plt.suptitle('Subplots',fontsize = 20)

#first
plt.subplot(1,2,1) #(number of rows, number of columns, number of plot)
plt.pie(RP,labels=Teams, autopct='%1.2f%%', colors =colors,  startangle=180, radius=1.5)
plt.title('Rank Points');

#second
plt.subplot(1,2,2)
plt.scatter(DFrame['Rank_Points'],DFrame['Year'])
plt.title('Rank points vs Years: ')
plt.xlabel('Rank Points')
plt.ylabel('Years'); 

#third 
plt.subplot(1,2,1)
Teams = ['Bangladish','England','Australia','India','Srilanka']
AN= [38000,40000,45000,55000,88000]
colors = ['green','blue']
plt.figure(figsize=(20,10))
plt.bar(Teams, AN, color= colors)
plt.title('Top Team Audience',fontsize = 20)
plt.xlabel('Teams',fontsize = 20)
plt.ylabel('Audience Number',fontsize = 20)
plt.show()



```
